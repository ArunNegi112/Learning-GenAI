LLMs hava parameterized knowledge - their knowledge is stored in the form of parameters 

Limitations of LLMs: 
- Do not have knowledge about private data, hence cannot answer queries based on private data on which model is not trained
- Knowledge is only limited to date the model is trained 
- Hallucinations: Give wrong answer with high confidence 

To solve these limitations we can do something called "Fine-tuning":
- Gather data -carefully curated examples
- Choose a method - full parameter fine tuning, LoRaA/Q-LoRA etc
- Train the model again for few epochs - You keep the base parameters frozen/partially-frozen and update only small subset of parameters
- Evaluate and run safety test

But obv Fine-tuning this has limitations too
- Training LLMs is very costly 
- You may have to pretrain it again and again to keep the model always updated 
- Very complex and time taking 

To overcome these limitations we have "In-context Learning":
- Here the model learns purely through examples given in the prompt - without updating its weights 

Now as LLMs got far better, what if instead of just few shot prompting, we inject the whole context about the query - documents, facts, chat history, background information etc 
This is called "RAG" Retrieval Augmented Generation

4 steps in RAG:
- Indexing: 
    - Its the process of preparing/storing knowledge or data such search is efficient
    - This has 4 steps:
        - Document Ingestion: Document loading
        - Text Chunking: Text splitting 
        - Embeddings: Similar vectors land close in vector space hence semantic search is possible
        - Storage in vector store

- Retrieval:
    - Its the process of finding most relevant information from pre-built index based on user's query
    - For this we use many types of semantic searches

- Augmentation:
    - In this step we combine the retrieved information, instructions, and users query to form a new, better prompt for LLM
