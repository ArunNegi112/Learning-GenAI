We know how to build a basic RAG based application - Below are the improvements to be done to make it production ready

1. UI Improvements 
    - Using HTML/CSS etc of webdev or For simplicity - Streamlit

2. Evaluation 
    - We evaluate the RAG system based on multiple parameters. 
        - Ragas: evaluate based on parameters like: Faithfulness, context precision, context recall, answer relevancy
        - Langsmith: used for tracing

3. Indexing 
    - Improve the preprocessing of data, Text splitting, Embeddings, Vector Stores

4. Retrieval
    - a. Pre-Retrieval
        - Query Rewriting: Rewrite the query using llm and then retrieve docs based on new, better query
        - Multiple Query Generation: Use llm to generate multiple similar queries and then perform retrieval
        - Domain aware routing: Multiple retrievers in the system, so we send query to different retrievers based on type of query
    
    - b. During retrieval
        - MMR
        - Hybrid retrieval: similarity + keyword matching
    
    - c. Post retrieval 
        - Contextual compression: Only keep the meaningful part from retrieved docs in the further steps
