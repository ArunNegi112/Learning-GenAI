## What is LangChain?
LangChain is an open source framework for building applications powered by _Foundation Models_ (LLMs or LMMs).   

### Why LangChain? 
- It supports almost all the available LLMs 
- It provides an interface through which we can use any LLM model without any much changes in code
- It has a concept of Chains, chains can convert complex workflows into a simple pipeline
- Highly available resources since its open source
- Provides complete ecosystem for any kind of application, ex: many types of _doc loaders_, _embeddings_ etc
