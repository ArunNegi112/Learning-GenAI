This a small project where im going to use `cosine similarity` to retrieve the corresponding information from documents about a question that user asks

```python
from sklearn.metrics.pairwise import cosine_similarity
from langchain_huggingface import HuggingFaceEmbeddings 
#You should use openai models but yeah fu.k that shit im broke
```

```python
documents = [
    "Virat Kohli is an Indian cricketer known for his aggressive batting and leadership.",
    "MS Dhoni is a former Indian captain famous for his calm demeanor and finishing skills.",
    "Sachin Tendulkar, also known as the 'God of Cricket', holds many batting records.",
    "Rohit Sharma is known for his elegant batting and record-breaking double centuries.",
    "Jasprit Bumrah is an Indian fast bowler known for his unorthodox action and yorkers."
]

query = 'tell me about kohli'
```

```python
embedding_model = HuggingFaceEmbeddings(model = "sentence-transformers/all-MiniLM-L6-v2")

documents_vector = embedding_model.embed_documents(documents)
query_vector = embedding_model.embed_query(query)
```

```python
similarity_scores = cosine_similarity(X=documents_vector,Y=[query_vector])
#The inputs inside cosine_similarity should be list of 2 dimensions

similarity_scores
```

```python
import numpy as np

index = int(np.where(similarity_scores == similarity_scores.max())[0][0])
max_score = float(similarity_scores.max())

print(query)
print(documents[index])
print("Similarity score: ",max_score)
```
