## 1. Models
Models are interfaces through which you interact with any foundation model  
And langchain have made this interface standardised for all models

Example, calling an api through googles model and anthropics model

``` python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv

load_dotenv()  
google_model = ChatGoogleGenerativeAI(model='gemini-2.5-flash')
response = google_model.invoke("What is chemical formula for water")
print(response.content)

``` python
from langchain_anthropic import ChatAnthropic

anthropic_model = ChatAnthropic(model_name='claude-3-opus-20240229')
response = anthropic_model.invoke("What is the chemcial formula of water")
print(response.content)

### Two types of Models
1. LLMs (That take text input and return text output)
2. Embedding models

- Embedding models are models that convert input text to a vector of fixed length in a higher dimensional space. This is useful in _semantic search_ 

- Semantic search is a search approach that understands the actual meaning and intent of input and then search for output instead of just matching the keywords

For now we will use gemini, since that is free till a limit

## 5. Memory

Problem: LLM api calls are _stateless_, that means, llm does not remember its previous api call   
For that LangChain provides a highly useful component _Memory_. 


Types of memory:   
- _ConversationBufferMemory_: stores a transcript of previous messages and give all of it as input in next api call
- _ConversationBufferWindowMemory_: Only stores last n number of interaction
- _Summarizer-based memory_ : Summarizes the previous chats and then provide it as input in next api call
- _Custom Memory_: you can store specialised states such as user perferences, user age etc. in a custom memory class
