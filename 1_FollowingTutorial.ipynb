### Thinking of models
Gemini 2.5 series models use thinking by default internally for better answers and solving complex problems.  
- If my problem is complex and i want model to think more i can increase the `thinkingBudget` 
- But it costs more obviously, if my problem is not complex and i want to increase my latency i can set `thinkingBudget = 0` or   
- If i want _dynamic thinking_ (model understands when to think according to problem itself) i can set `thinkingBudget = -1` 

Below is the range of thinkingBudget and its value if we want no thinking or dynamic thinking

```python
from google.genai import types

client = genai.Client()

response = client.models.generate_content(
    model= 'gemini-2.5-flash',
    contents= prompt,
    config= types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(thinking_budget=0)  #turn off thinking
    )
)

print(response.text)
```
